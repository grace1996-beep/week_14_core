# Feature selection

**Method 1 Filter method**

```python
%%R
# Installing and loading caret package
# ---
# 
suppressWarnings(
        suppressMessages(if
                         (!require(caret, quietly=TRUE))
                install.packages("caret")))
library(caret)
```

```python
%%R
# Installing and loading the corrplot package for plotting
# ---
# 
suppressWarnings(
        suppressMessages(if
                         (!require(corrplot, quietly=TRUE))
                install.packages("corrplot")))
library(corrplot)
```

```python
%%R
# Selecting the numerical data (excluding the categorical variables)
# ---
# 
df <- df3[,c(6,8,12,13,14,15,16)]
head(df)
```

```python
%%R
# Calculating the correlation matrix
# ---
#
correlationMatrix <- cor(df)
```

```python
%%R
# Find attributes that are highly correlated
# ---
#
highlyCorrelated <- findCorrelation(correlationMatrix, cutoff=0.75)

# Highly correlated attributes
# ---
# 
highlyCorrelated

names(df[,highlyCorrelated])
```

```python
%%R
# I can remove the variables with a higher correlation 
# and comparing the results graphically as shown below
# ---
# 
# Removing Redundant Features 
# ---
# 
df2 <- df[-highlyCorrelated]
```

```python
%%R
#performing graphical comparison
par(mfrow = c(1, 2))
corrplot(correlationMatrix)
corrplot(cor(df2))
```
